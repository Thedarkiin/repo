```python
# Fill numeric columns with median
for col in num_cols:
    med = df_train[col].median()
    df_train[col].fillna(med, inplace=True)
    df_holdout[col].fillna(med, inplace=True)

# Fill categorical columns with mode
for col in cat_cols:
    mode = df_train[col].mode()[0]
    df_train[col].fillna(mode, inplace=True)
    df_holdout[col].fillna(mode, inplace=True)
```

```python
import sys
sys.path.insert(0, "..")

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from src.modeling import train_test_data, build_models

# Ensure output directory exists
Path("reports/figures").mkdir(parents=True, exist_ok=True)

# 1. Split data
X_train, X_test, y_train, y_test = train_test_data(df, selected)

# 2a. Prepare ROC comparison figure
fig_roc, ax_roc = plt.subplots(figsize=(8,6))

# 2b. Train & evaluate each model
for name, mdl in build_models().items():
    mdl.fit(X_train, y_train)
    y_pred = mdl.predict(X_test)
    y_proba = mdl.predict_proba(X_test)[:,1]
    print(f"Running: {name}")

    
    # Classification report
    print(f"=== {name} ===")
    print(classification_report(y_test, y_pred, digits=3))
    
    # Confusion matrix plot
    cm = confusion_matrix(y_test, y_pred)
    fig_cm, ax_cm = plt.subplots()
    im = ax_cm.imshow(cm, cmap="Blues")
    ax_cm.set_title(f"{name} Confusion Matrix")
    ax_cm.set_xticks([0,1]); ax_cm.set_yticks([0,1])
    ax_cm.set_xticklabels(["Stay","Churn"]); ax_cm.set_yticklabels(["Stay","Churn"])
    for i in (0,1):
        for j in (0,1):
            ax_cm.text(j, i, cm[i,j], ha="center", va="center", color="white")
    fig_cm.savefig(f"reports/figures/{name.lower()}_cm.png")
    plt.close(fig_cm)
    
    # Add this modelâ€™s ROC to the ROC figure
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    roc_auc = auc(fpr, tpr)
    ax_roc.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")

# 3. Finalize & save ROC comparison
ax_roc.plot([0,1],[0,1], linestyle="--", color="gray")
ax_roc.set_xlabel("False Positive Rate")
ax_roc.set_ylabel("True Positive Rate")
ax_roc.set_title("ROC Curve Comparison")
ax_roc.legend(loc="lower right")
fig_roc.savefig(r"C:\Users\aserm\OneDrive\Bureau\telecomchurn\reports\figures\roc_comparison.png")
plt.close(fig_roc)

print("Saved confusion matrices and ROC comparison to reports/figures/")
```
